PROJECT 1: "Area-Price Prediction Model"

Description: An Area-Price Prediction ML model is a machine learning model designed to forecast real estate prices based on various factors related to specific geographical areas. By analyzing historical sales data, location attributes (like proximity to amenities, crime rates, school quality), and economic indicators, the model aims to provide accurate predictions of property prices in different neighborhoods or regions. This enables stakeholders such as buyers, sellers, and investors to make informed decisions regarding real estate transactions.

```python
#Importing Python Third party Libraries:
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Reading the csv file:
df = pd.read_csv("Housing.csv")
```

```python
#Splitting the data into Independent and Dependent Variables:
x = df[["area"]]
y = df.price
```

```python
#Visualizing the data by using Scatter Plot:
'''
plt.title("Area vs Price")
plt.xlabel("Area(in sq.m)")
plt.ylabel("Price(in Rs.)")
plt.scatter(x,y)
plt.show()
'''
```

```python
#Analyzing the scatter plot with the help of Linear Regression model by fitting the line:

model = LinearRegression()
model.fit(x,y)

prediction = model.predict(x)

plt.title("Area vs Price")
plt.xlabel("Area(in sq.m)")
plt.ylabel("Price(in Lakhs Rs.)")
plt.scatter(x, y, color="red", ec="black", marker="o")
plt.plot(x, prediction, color="blue")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
#Asking from user and predicting the price of the area:
myArea = int(input("Enter the area: "))
myPrice = model.predict([[myArea]])
print("Price is: ", myPrice)
```

```python
#Checking for the coefficient and intercept values:
model.coef_
model.intercept_
```
